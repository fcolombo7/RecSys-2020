```python
block_size = int(len(profile_length)*0.15)
block_size
```

```python
sorted_users = np.argsort(profile_length)
```

```python
for group_id in range(0, 7):
    
    start_pos = group_id*block_size
    end_pos = min((group_id+1)*block_size, len(profile_length))
    
    users_in_group = sorted_users[start_pos:end_pos]
    
    users_in_group_p_len = profile_length[users_in_group]
    
    print("Group {}, average p.len {:.2f}, min {}, max {}, count{}".format(group_id, 
        users_in_group_p_len.mean(), users_in_group_p_len.min(), users_in_group_p_len.max(), len(users_in_group_p_len)))
```

```python
from Base.Evaluation.Evaluator import EvaluatorHoldout
MAP_05_per_group = []
MAP_06_per_group = []
MAP_UW_per_group = []
MAP_icb_per_group = []
cutoff = 10

for group_id in range(0, 6):
    
    start_pos = group_id*block_size
    end_pos = min((group_id+1)*block_size, len(profile_length))
    
    users_in_group = sorted_users[start_pos:end_pos]
    
    users_in_group_p_len = profile_length[users_in_group]
    
    print("Group {}, average p.len {:.2f}, min {}, max {}".format(group_id, 
        users_in_group_p_len.mean(), users_in_group_p_len.min(), users_in_group_p_len.max()))
    
    
    users_not_in_group_flag = np.isin(sorted_users, users_in_group, invert = True)
    users_not_in_group = sorted_users[users_not_in_group_flag]
    
    evaluator_test = EvaluatorHoldout(URM_test, cutoff_list=[cutoff], ignore_users = users_not_in_group)
    
    
    results, _ = evaluator_test.evaluateRecommender(recommender5)
    MAP_05_per_group.append(results[cutoff]["MAP"])
 
    results, _ = evaluator_test.evaluateRecommender(recommender6)
    MAP_06_per_group.append(results[cutoff]["MAP"])
    
    #results, _ = evaluator_test.evaluateRecommender(rec_uw)
    #MAP_UW_per_group.append(results[cutoff]["MAP"])
    
    results, _ = evaluator_test.evaluateRecommender(itemKNNCBF)
    MAP_icb_per_group.append(results[cutoff]["MAP"])
```

```python
import matplotlib.pyplot as pyplot
%matplotlib inline  

pyplot.plot(MAP_05_per_group, label="Lin05")
pyplot.plot(MAP_06_per_group, label="Lin06")
pyplot.plot(MAP_icb_per_group, label="ItemCBF")
pyplot.ylabel('MAP')
pyplot.xlabel('User Group 20%')
pyplot.legend()
pyplot.show()
```
